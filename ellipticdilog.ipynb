```gp
\\ Using GPT-5.2, updated the file "elldilog.gp" that was originally available on 
\\ https://sweet.ua.pt/apacetti/gp/elldilog.gp.txt (Rodriguez Villegas)
\\
\\
\\ Elliptic dilog of a POINT P on E (complex case).
\\ Uses Zagier single-valued dilogarithm: polylog(2, *, 2).
elldilog(E, P)=
{
  my(om, w1, w2, u, tau, z, q, qq, s, N, log10q);

  om = ellperiods(E);       \\ returns [w1,w2] 
  w1 = om[1];  w2 = om[2];

  u   = ellpointtoz(E, P) / w1;      \\ complex elliptic logarithm modulo lattice
  tau = w2 / w1;

  z = exp(2*Pi*I*u);
  q = exp(2*Pi*I*tau);

  \\ Choose truncation N so |q|^N is well below working decimal precision
  log10q = log(abs(q)) / log(10);    \\ < 0
  N = max(10, ceil((precision(1.0)+5)/(-log10q)));

  qq = 1;
  s  = polylog(2, z, 2);
  for(n=1, N,
    qq *= q;
    s += polylog(2, qq*z, 2) + polylog(2, z/qq, 2);
  );
  s
}

addhelp(elldilog, "Compute elliptic dilog of point P on elliptic curve E.");

elllincomb(E, L)=
{
  my(R = [0]); \\ point at infinity
  for(i=1, #L,
    R = elladd(E, R, ellmul(E, L[i][2], L[i][1]));
  );
  R
}

elldilog_div(E, L)=
{
  my(s=0);
  for(i=1, #L,
    s += L[i][1] * elldilog(E, L[i][2]);
  );
  s
}
```
